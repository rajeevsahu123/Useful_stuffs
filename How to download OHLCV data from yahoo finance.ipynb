# How to download OHLCV data from yahoo using pandas-datareader
for more information [Pandas-datareader](http://pandas-datareader.readthedocs.io/en/latest/remote_data.html)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline

from pandas_datareader import data as pdr
import datetime

end = datetime.date.today()
begin=end-pd.DateOffset(365*5)

st=begin.strftime('%Y-%m-%d')
ed=end.strftime('%Y-%m-%d')

df = pdr.get_data_yahoo("AAPL",st,ed)
```

```python
df.describe()
```

# Visualization incomplete working on it

```python
fig = plt.figure(figsize=(10,8))
# Set the outer colour to white
fig.patch.set_facecolor('white') 

# Plot the equity curve
ax1 = fig.add_subplot(311, ylabel='Close')
df['Close'].plot(ax=ax1, color="blue", lw=2.)
plt.grid(True)

# Plot the returns
ax2 = fig.add_subplot(312, ylabel='Volume')
df['Volume'].plot(ax=ax2, color="black", lw=2.)
plt.grid(True)

# Plot the returns
ax3 = fig.add_subplot(313, ylabel='Max draw down, %')
#df['Adj Close'].plot(ax=ax3, color="red", lw=2.)
ax3.fill_between(df.index, df['Adj Close'], 200, color='red')
plt.grid(True)
```
